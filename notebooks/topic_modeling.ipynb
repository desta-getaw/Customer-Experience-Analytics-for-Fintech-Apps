```python
import sys
import os
sys.path.append(os.path.abspath("../scripts"))
from topic_modeling import TopicModeler
import pandas as pd

# Load preprocessed reviews (or original review text)
df = pd.read_csv("../data/cleaned_reviews.csv")
documents = df["review"].dropna().tolist()

# Instantiate and run LDA or NMF
topic_modeler = TopicModeler(n_topics=60, model_type="lda")
model, vectorizer = topic_modeler.fit(documents)

# Display topics
topics_df = topic_modeler.display_topics(n_words=8)
display(topics_df)

# Optionally save results
import os
os.makedirs("../data/outputs", exist_ok=True)
topics_df.to_csv("../data/outputs/review_topics.csv", index=False)
```
